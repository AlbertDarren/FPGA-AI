# 图像处理初步-亮度处理

```python
# 灰度值
value=150
```

```python
def pixel_value(pixel:int,k:int=8):
    """
    adjust the origin pixel value to a proper value in the range [0,255]
    pixel:the origin pixel value
    k:the depth of the origin image,default 8
    return:a proper pixel value in the range [0,255]
    """
    upper_limit,lower_limit=(1<<k)-1,0
    new_pixel_value=pixel
    if pixel>upper_limit:
        new_pixel_value=upper_limit
    elif pixel<lower_limit:
        new_pixel_value=lower_limit
    return new_pixel_value
```

```python
# 将pixel_value函数转换为对数组中每个元素的ufunc函数
pixel_value_vectorize=np.vectorize(pixel_value,otypes=[np.uint8])
bright_g_img=pixel_value_vectorize(img[:,:,1]+value)
plt.imshow(bright_g_img,cmap="gray")
```
